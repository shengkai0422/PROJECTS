```python
import pandas
dfs=pandas.read_html('https://www.cht.com.tw/home/apple/iphone/index')
```

```python
df=dfs[0]
```

```python
df.head()
```

```python
#extrat those row that contains info "個月"
df[df['月繳金額.1'].str.contains("個月")]
```

```python
#regular expression 
df['month']=df['月繳金額.1'].str.extract('(\d+)個月')
```

```python
df.head()
```

```python
#fill the month column 
df['month']=df[ 'month'].ffill()
```

```python
df.head()
```

```python
#deleat th NaN lines
df.dropna(inplace=True)
```

```python
df.head()
```

```python
#exclude "個月" inofrmation lines
df2=df[~df['月繳金額.1'].str.contains('個月')]
```

```python
df2.head()
```

```python
df2.columns
```

```python
df2.columns=['月繳金額', '机型', '容量', '$599', '$799', '$999', '$1199', '$1,399',
       '$1,599', '$1,799', '$2,699', 'month']
```

```python
df2.head()
```

```python
df3=pandas.wide_to_long(df2,['$'],i=['月繳金額', '机型', '容量','month'],j='monthly_price',suffix='[\d,]+')
```

```python
 df3.head()
```

```python
df3.reset_index(inplace=True)
```

```python
df3.head()
```

```python
df4=df3[df3['$']!='-']
```

```python
df4.head()
```

```python
df4.info()
```

```python
#conversion type 
df4['month']=df4['month'].astype(int)
```

```python
df4['monthly_price']=df4['monthly_price'].map(lambda e: int(e.replace(',','')))
```

```python
df4['$']=df4['$'].astype(int)
```

```python
df4.head()
```

```python
df4['total_price']=df4['$']+df4['month']*df4['monthly_price']
```

```python
df4.head()
```

```python
df4.sort_values('total_price')
```
